### Handling categorical data with Label encoder

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
for i in cat:
    data[i] = le.fit_transform(data[i])
```

### Spliting the data for regression and classification

### Classification

```python
X_c = data.drop('Default',axis=1)
y_c = data[['Default']]
```

```python
X_c
```

```python
y_c
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier,DecisionTreeRegressor
from sklearn.metrics import accuracy_score,r2_score
```

```python
c_X_train,c_X_test,c_y_train,c_y_test = train_test_split(X_c,y_c,test_size=0.2)
```

```python
dtc = DecisionTreeClassifier()
```

```python
dtc.fit(c_X_train,c_y_train)
```

```python
c_y_pred = dtc.predict(c_X_test)
```

```python
accuracy_score(c_y_pred,c_y_test)
```
