### Label encoder

```python
from sklearn.preprocessing import LabelEncoder
```

```python
#separeting categorical data
cat = data.columns[data.dtypes=='object']
```

```python
#separeting numarical data
num = data.select_dtypes(include=['float64','int64']).columns
```

### Correlation Visual

### Splitting feature and target data for Classification and Regression

```python
X_cls = data.drop('y',axis=1)
y_cls = data['y']
X_reg = data.drop('duration',axis=1)
y_reg = data['duration']
```

### Train test split

```python
from sklearn.model_selection import train_test_split
```

```python
#classification algorithm
X_train,X_test,y_train,y_test = train_test_split(X_cls,y_cls,test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier(n_estimators=100)
```

```python
rfc.fit(X_train,y_train)
```

```python
rfc_prd = rfc.predict(X_test)
```

### Confusion matrix

```python
from sklearn.metrics import confusion_matrix
```

```python
cm = confusion_matrix(y_test,rfc_prd)
```

```python
cm
```

```python
#regression algorithm
X_train,X_test,y_train,y_test = train_test_split(X_reg,y_reg,test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rr = RandomForestRegressor(n_estimators=100)
```

```python
rr.fit(X_train,y_train)
```

```python
rr_pred = rr.predict(X_test)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test,rr_pred)
```
